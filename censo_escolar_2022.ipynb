```python
censo_2022_pb_estadual = censo_2022.loc[(censo_2022['SG_UF'] == 'PB') & (censo_2022['TP_DEPENDENCIA'] == 2) & (censo_2022['TP_SITUACAO_FUNCIONAMENTO'] == 1)]
```

```python
censo_2022_pb_estadual_conectividade = censo_2022_pb_estadual[['IN_INTERNET', 'IN_INTERNET_ALUNOS', 'IN_INTERNET_ADMINISTRATIVO', 'IN_INTERNET_APRENDIZAGEM', 'IN_INTERNET_COMUNIDADE', 'IN_ACESSO_INTERNET_COMPUTADOR', 'IN_ACES_INTERNET_DISP_PESSOAIS', 'TP_REDE_LOCAL', 'IN_BANDA_LARGA']]
censo_2022_pb_estadual_conectividade.reset_index(inplace=True, drop=True)
censo_2022_pb_estadual_conectividade.head()
```

```python
censo_2022_pb_estadual_conectividade = censo_2022_pb_estadual_conectividade.rename({'IN_INTERNET':'acesso_internet', 'IN_INTERNET_ALUNOS':'interntet_para_uso_dos_alunos', 'IN_INTERNET_ADMINISTRATIVO':'internet_para_uso_administrativo', 'IN_INTERNET_APRENDIZAGEM': 'internet_para_aprendizagem', 'IN_INTERNET_COMUNIDADE':'internet_para_uso_da_comunidade', 'IN_ACESSO_INTERNET_COMPUTADOR':'alunos possuem_tablet_computadores_na_escola', 'IN_ACES_INTERNET_DISP_PESSOAIS':'acesso a internet_disponivel_para_disposit', 'TP_REDE_LOCAL':'tipo_rede_local', 'IN_BANDA_LARGA':'possui_internet_banda_larga'}, axis='columns' )
```

```python
censo_2022_pb_estadual_conectividade.head()
```

```python
censo_2022_pb_estadual_conectividade['acesso_internet'].value_counts(ascending=True)
```

```python
print(f'Percentual das escolas que possuem acesso a internet(Censo de 2022) {censo_2022_pb_estadual_conectividade["acesso_internet"].value_counts(ascending=True, normalize=True)[1].round(4)*100}%')
```

```python
print(f'Percentual das escolas que possuem acesso do tipo banda larga(Censo de 2022) {censo_2022_pb_estadual_conectividade["possui_internet_banda_larga"].value_counts(ascending=True, normalize=True)[1].round(4)*100}%')
```
